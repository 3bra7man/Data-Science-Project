```python
#first dataset
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')
data.shape
```

```python
plt.figure()
values = {'nr records': data.shape[0], 'nr variables': data.shape[1]}

plt.bar(values.keys() , values.values())
plt.title('Nr of records vs nr variables')

plt.figure(figsize=(4,2))

plt.show()
```

```python
data.dtypes #All datatypes are numeric so no need to cast anything
```

```python
plt.figure()
mv = {}
for var in data:
    mv[var] = data[var].isna().sum()
plt.bar(mv.keys(), mv.values())
plt.title('Nr of missing values per variable')
plt.xlabel('variables')
plt.ylabel('nr missing values')
mv.values()
#no missing values
```

```python
#Second dataset
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
data = pd.read_csv('qsar_oral_toxicity.csv' , lineterminator='\n')
data
```

```python
plt.figure()
values = {'nr records': data.shape[0], 'nr variables': data.shape[1]}

plt.bar(values.keys() , values.values())
plt.title('Nr of records vs nr variables')

plt.figure(figsize=(4,2))

plt.show()
```

```python
data.dtypes
```

```python
cat_vars = data.select_dtypes(include='object')
data[cat_vars.columns] = data.select_dtypes(['object']).apply(lambda x: x.astype('category'))
data.dtypes
```

```python
plt.figure()
mv = {}
for var in data:
    mv[var] = data[var].isna().sum()
plt.bar(mv.keys(), mv.values())
plt.title('Nr of missing values per variable')
plt.xlabel('variables')
plt.ylabel('nr missing values')
mv.values()
#no missing values
```
